```python
training_accuracy = []
val_accuracy = []
training_f1 = []
val_f1 = []
k_neighbors = []
for i in range (1,10):
   model = KNeighborsClassifier(n_neighbors= i, weights = 'uniform', p = 3)
   model.fit(X_train, y_train)
   y_pred = model.predict(X_train)
   training_accuracy.append(accuracy_score(y_train, y_pred))
   val_accuracy.append(cross_val_score(model, X_train, y_train, cv = 5, scoring = 'accuracy').mean())
   training_f1.append(f1_score(y_train, y_pred))
   val_f1.append(cross_val_score(model, X_train, y_train, cv= 5, scoring='f1_macro').mean())
   k_neighbors.append(i)
K_neighbors = {"Training Accuracy": training_accuracy, "Validation Accuracy": val_accuracy, "Training F1": training_f1, "Validation F1":val_f1, "K_neighbors": k_neighbors}
K_neighbors_df = pd.DataFrame.from_dict(K_neighbors)
```

```python
plot_df = K_neighbors_df.melt("K_neighbors",var_name='Metrics',value_name="Values")
fig,ax = plt.subplots(figsize=(15,5))
sns.pointplot(x="K_neighbors", y="Values",hue="Metrics", data=plot_df,ax=ax)
```
