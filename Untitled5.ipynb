```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/chau25102001/myrep/main/new_bank.csv")
data.head()
data.columns
data = data.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
data
```

```python
data['education'].unique()
```

```python
def tran_education(x):
  if x == 'unknown':
    return -1
  if x == "primary":
    return 1
  if x == "secondary":
    return 2
  if x== 'tertiary':
    return 3
  else:
    return x
def tran_binary(x):
  if x == 'yes':
    return 1
  if x =='no':
    return 0
  else:
    return x
```

```python
pd.get_dummies(data)
```

```python
data = pd.get_dummies(data)
data
```

```python
list1 = data['pdays'].values
count = 0
for i in list1:
  if i == -1:
    count += 1
count
```

```python
data['education'] = data['education'].apply(tran_education)
data['default'] = data['default'].apply(tran_binary)
data['housing'] = data['housing'].apply(tran_binary)
data['loan'] = data['loan'].apply(tran_binary)
```
